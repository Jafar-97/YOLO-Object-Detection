```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

## Make Detections with Images

https://pytorch.org/hub/ultralytics_yolov5/

```python
# loding pritrain model using torch hub
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
results = model('bus.jpg')
results.print()
```

```python
%matplotlib inline
plt.imshow(np.squeeze(results.render()))
plt.show()
```

```python
results = model('https://upload.wikimedia.org/wikipedia/commons/9/9e/Giraffe_Mikumi_National_Park.jpg')
results.print()
```

```python
%matplotlib inline
plt.imshow(np.squeeze(results.render()))
plt.show()
```
